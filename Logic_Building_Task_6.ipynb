1. Smart Parking Lot Management System

```python
def manage_parking(capacity, logs):
    curr_count=0
    peak_usage=0
    for action in logs:
        if action=="IN":
            curr_count+=1 # Count of vehicle entry
        elif action=="OUT":
            curr_count=max(0, curr_count-1) # Count of vehicle exit
        if curr_count>peak_usage:
            peak_usage=curr_count
    if curr_count>=capacity:
        status="Full / Alert: Capacity Exceeded"
    elif curr_count>=capacity*0.9:
        status="Nearly Full"
    else:
        status="Available"
    print(f"Currently Parked Vehicles: {curr_count}")
    print(f"Peak Usage Today: {peak_usage}")
    print(f"Parking Status: {status}")
capacity_limit=50
vehicle_logs=["IN", "IN", "IN", "OUT", "IN", "IN", "IN", "OUT"]
manage_parking(capacity_limit, vehicle_logs)
```

2. Online Food Delivery Time Estimator

```python
def estimate_delivery(distance, traffic, weather):
    base_time=distance*5 # Base Time with 5mins time for 1km
    traffic_delay=0
    if traffic=="Medium": # Delay due to traffic
        traffic_delay=10
    elif traffic=="High":
        traffic_delay=20
    subtotal_time=base_time+traffic_delay
    if weather=="Rainy": # Delay due to weather condition
        final_eta=subtotal_time+10
    elif weather=="Stormy":
        final_eta=subtotal_time+20
    else:
        final_eta=subtotal_time
    return f"Estimated Delivery Time: {int(final_eta)} minutes"
print(estimate_delivery(12, "Medium", "Stormy"))
```

3. Movie Theatre Seat Occupancy Analyzer

```python
def analyze_theatre_occupancy(total_seats, booked_list):
    num_booked=len(booked_list) # To Count number of '1's in list
    occupancy_pct=(num_booked/total_seats)*100 # To calculate occupancy percentage
    if occupancy_pct==100: # To determine status and display suggestion
        status="Housefull"
        suggestion="Open an extra show immediately."
    elif occupancy_pct>=80:
        status="Almost Full"
        suggestion="Consider scheduling an additional show."
    elif occupancy_pct>=50:
        status="Filling Fast"
        suggestion="Keep monitoring ticket sales."
    else:
        status="Available"
        suggestion="Plenty of seats left."
    print(f"Occupancy: {int(occupancy_pct)}%")
    print(f"Show Status: {status}")
    print(f"Action: {suggestion}")
total_capacity=250
bookings=[1]*220
analyze_theatre_occupancy(total_capacity, bookings)
```

4. Cloud Server Load Classification System

```python
def classify_server_load(readings):
    if not readings:
        return "No data available"
    avg_cpu=sum(readings)/len(readings)
    if avg_cpu>=80:
        status="Critical"
        alert="ALERT: Immediate scaling or optimization required!"
    elif 50<=avg_cpu<80:
        status="Warning"
        alert="NOTE: Resources are heavily utilized."
    else:
        status="Normal"
        alert="SYSTEM: Operating within safe parameters."
    print(f"Average CPU Load: {int(avg_cpu)}%")
    print(f"Server Status: {status}")
    print(f"System Message: {alert}")
cpu_readings=[46, 60, 78, 82, 95]
classify_server_load(cpu_readings)
```

5. Smart Classroom Resource Usage Monitor

```python
def monitor_class_rsc(usage_dict):
    thresholds={ # Define max hours allowed for each resource
        "Projector": 5,
        "AC": 8,
        "Lights": 10
    }
    overused=[]
    energy_alert="No"
    for resource, hours in usage_dict.items(): # To check usage against limit
        if resource in thresholds:
            if hours>thresholds[resource]:
                overused.append(resource)
                energy_alert="Yes"
    overused_str = ", ".join(overused) if overused else "None"
    print(f"Overused Resources: {overused_str}")
    print(f"Energy Alert: {energy_alert}")
current_usage = {
    "Projector": 8,
    "AC": 10,
    "Lights": 6
}
monitor_class_rsc(current_usage)
```

6. Online Event Registration Capacity Controller

```python
def mng_event_regist(capacity, total_registrations):
    confirmed=0
    waitlisted=0
    status=""
    if total_registrations>capacity:
        confirmed=capacity
        waitlisted=total_registrations-capacity
        status="Closed (Waitlist Active)"
    else:
        confirmed=total_registrations
        waitlisted=0
        status="Open"
    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")
mng_event_regist(100, 105)
```
