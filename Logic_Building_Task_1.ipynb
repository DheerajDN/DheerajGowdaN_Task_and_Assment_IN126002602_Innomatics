1. User Login Check

```python
user_input=input("Enter Username : ") # user input
pass_input=input("Enter Password : ") # user input

if user_input=="admin" and pass_input=="1234": # Checking Logic
  print("Login Successful") # If both the condition are true this statement is printed
else:
  print("Invalid Credentials") # If either Username/Password is wrong this statement is printed
```

2. Pass/Fail Analyzer

```python
marks=[45,78,90,33,60] # Given Students Marks

p_count=0 # counters are initialized to zero
f_count=0

for score in marks: # Logic to check if the score meets the passing criteria
  if score>=50:
    p_count+=1
  else:
    f_count+=1

print("Total Students :",len(marks))
print("Passed Students : ",p_count)
print("Failed Students : ",f_count)
```

3. Simple Data Cleaner

```python
names=[" Alice ","bob"," CHARLIE "] # Given Names

Clean_names=[] # Empty List to store the cleaned names

for name in names: # Logic to apply required string methods
  C_names=name.strip().lower()
  Clean_names.append(C_names)

print("Original List", names)
print("Cleaned List", Clean_names)
```

4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"] # Given Message

for msg in messages:
  length=len(msg) # To Calculate Message Length using len
  print(f"Message: '{msg}' , Length: {length}")

  if length>10: # To check if the message length is exceeding the limit
    print("---->FLAGGED: Message too long (exceeded message length limit :10)")
```

5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"] # Given System Logs

err_count=0

for entry in logs:
  if entry == "ERROR":
    err_count+=1
print("Total Error found in the logs :", err_count)
```
