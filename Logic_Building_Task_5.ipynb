1. Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
    total_likes = 0
    for likes in likes_list: # To Calculate Total likes
        total_likes+=likes
    if total_likes>=1000: #  To Check if the Post is viral or not
        status="Viral Post"
    else:
        status="Normal Engagement"
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")
post_likes=[400, 350, 200, 100]
analyze_engagement(post_likes)
```

3. Agriculture – Rainfall Adequacy Checker

```python
def check_rainfall(rainfall_data, required_level):
    total_rainfall=0
    count=0
    for amount in rainfall_data:
        total_rainfall+=amount # To calculate total rain fall
        count+=1
    average_rainfall=total_rainfall/count # To calculate average rainfall
    if average_rainfall>=required_level:
        status="Adequate Rainfall"
    else:
        status="Inadequate Rainfall"
    print(f"Average Rainfall: {average_rainfall}")
    print(f"Rainfall Status: {status}")
monthly_rain=[80, 65, 90, 53]
required=70
check_rainfall(monthly_rain, required)
```

4. Social Media – Duplicate Account Detection

```python
def detect_duplicates(usernames):
    unique_usernames=set(usernames)
    if len(unique_usernames)<len(usernames): # To check if dupliactes exist
        result="Yes"
    else:
        result="No"
    print(f"Duplicate Accounts Found: {result}")
user_list=["@user1", "@dev_pro", "@user1", "@pixel_art"]
detect_duplicates(user_list)
```

6. Agriculture – Premium Crop Price Filter

```python
def filter_premium_crops(crop_prices):
    premium_threshold=2000
    premium_list=[]
    regular_list=[]
    for price in crop_prices: # To check if the crops are premium or regular
        if price>premium_threshold:
            premium_list.append(price)
        elif price<premium_threshold:
            regular_list.append(price)
    print(f"Premium Crops: {premium_list}")
    print(f"Regular Crops: {regular_list}")
all_prices=[1500, 2500, 1800, 3200, 2000]
filter_premium_crops(all_prices)
```

7. System Monitoring – Application Health

```python
def check_app_health(error_count):
    if error_count==0: # To determine status based on error count
        status="Healthy"
    elif error_count<=5:
        status="Minor Issues"
    else:
        status="Critical Issues"
    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")
check_app_health(7)
```

9. E-Learning – Student Attendance Eligibility

```python
def check_exam_eligibility(attendance_record):
    total_days=len(attendance_record)
    days_present=0
    for status in attendance_record: # To calculate how many days student was present
        if status==1:
            days_present+=1
    attendance_percentage=(days_present/total_days) * 100
    if attendance_percentage>=75: # To check eligibility based on addendance
        eligibility="Eligible"
    else:
        eligibility="Not Eligible"
    print(f"Attendance Percentage: {attendance_percentage}")
    print(f"Exam Eligibility: {eligibility}")
attendance_data=[1, 1, 1, 0, 1]
check_exam_eligibility(attendance_data)
```

10. Smart Electricity Bill Analyzer

```python
def calculate_electricity_bill(units):
    total_bill=0
    if units<=100: # To determine rate per unit based on units used
        total_bill=units * 3
    elif units<=200:
        total_bill=(100 * 3)+((units-100) * 5)
    else:
        total_bill=(100 * 3)+(100 * 5)+((units-200) * 7)
    if total_bill<500: # To determine the status based on units used
        usage_status="Low Usage"
    elif 500<=total_bill<=1500:
        usage_status="Moderate Usage"
    else:
        usage_status="High Usage"

    return total_bill, usage_status
units_consumed = 250
bill, status = calculate_electricity_bill(units_consumed)
print(f"Units Consumed: {units_consumed}")
print(f"Total Bill: ₹{bill}")
print(f"Usage Status: {status}")
```
