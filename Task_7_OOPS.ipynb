1. Smart Light Controller

```python
class SmartLight:
    def __init__(self, name):
        self.name=name
        self.is_on=False # Set the Default State
    def set_status(self, action):
        if action.upper()=="ON":
            self.is_on=True
        elif action.upper()=="OFF":
            self.is_on=False
    def display_status(self):
        status="ON" if self.is_on else "OFF"
        print(f"{self.name} is {status}")
my_light=SmartLight("Gaming Room Light")
my_light.set_status("ON")
my_light.display_status()
```

2. Employee ID Card System

```python
class Emp:
    def __init__(self, name, emp_id, department):
        self.name=name
        self.emp_id=emp_id
        self.department=department
    def display_id(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")
emp_name="Akshay" # Inputs
id_num="EMP110"
dept="IT"
new_employee=Emp(emp_name, id_num, dept)
new_employee.display_id()
```

3. Mobile Contact Record

```python
class Contact:
    def __init__(self, name, phone):
        self.name=name
        self.phone=phone
        print("Contact Saved")
    def display_contact(self):
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")
contact_name="Bhuvan" # Inputs
ph_no="9875643201"
my_contact=Contact(contact_name, ph_no)
my_contact.display_contact()
```

4. Product Price Tag Generator

```python
class Product:
    def __init__(self, name, price):
        self.name=name
        self.price=price
    def display_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")
product_name="Gaming Headphones" # Input
product_price=5499
my_product=Product(product_name, product_price)
my_product.display_tag()
```

5. Movie Rating Display System

```python
class Movie:
    def __init__(self, name, rating):
        self.name=name
        self.rating=rating
    def display_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating}/5")
movie_name="Godfather" # Inputs
movie_rating=4.9
my_movie=Movie(movie_name, movie_rating)
my_movie.display_details()
```

6. Delivery Address Manager

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name=customer_name
        self.address=address
    def display_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")
name="Chethan"
loc="Pune"
my_delivery=Delivery(name, loc)
my_delivery.display_details()
```
