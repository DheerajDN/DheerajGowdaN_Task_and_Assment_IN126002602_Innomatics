1. Mobile Recharge Validation System

```python
def val_recharge():
    valid_plans = {199, 299, 399, 599} # Valid Plans

    while True:
        try:
            amount = int(input("Enter recharge amount : ₹ "))

            if amount < 50:
                print("Error: Recharge amount must be ≥ ₹50.")
            elif amount not in valid_plans:
                print(f"Error: Invalid plan. Choose from {valid_plans}.")
            else:
                print(f"Success: Recharge of ₹{amount}.")
                return amount

        except ValueError:
            print("Error: Please enter a valid number.")
val_recharge()
```

2. Inventory Reorder Alert System

```python
def check_inventory(inventory_data):
    for product, stock in inventory_data.items(): # Use Loop to go through the dictionary using .items() to get both the product and its stock
        if stock < 15:
            print(f"Reorder Alert : {product} (Current Stock : {stock})")
        else:
            print(f"Stock OK : {product} (Current Stock : {stock})")
if __name__ == "__main__":
    current_stock = {
        "Wireless Mice": 42,
        "Mechanical Keyboards": 10,
        "USB-C Cables": 8,
        "Laptop Stands": 25
    }
    check_inventory(current_stock)
```

3. Student Result Processing System

```python
def process_result(marks):
    if not marks:
        return "Fail"
    total=0
    for mark in marks:
        total +=mark
    average=total/len(marks)
    print("Average", average)

    if average >= 50:
        print("Pass")
        return "Pass"
    else:
        print("Fail")
        return "Fail"

if __name__ == "__main__":
    student_marks = [55, 62, 48, 70, 50]
    print("Student marks:", student_marks)
    process_result(student_marks)
```

4. Cab Fare Estimator with Retry Option

```python
def calculate_fare():
    while True:
        try:
            distance = float(input("Enter distance in km: "))
            if distance <= 0:
                print("Error: Distance cannot be negative.")
                continue
            peak=input("Is it peak hour? (yes/no): ")
            fare=50+(distance * 12)
            if peak=='yes':
                fare *= 1.25
            print(f"Estimated Fare: ₹{fare:.2f}")
            retry = input("Calculate another fare? (yes/no): ")
            if retry != 'yes':
                break
        except ValueError:
            print("Error: Please enter a valid numeric distance.")
calculate_fare()
```

5. Employee Attendance Eligibility

```python
def check_eligibility(attendance):
    if not attendance:
        return "Not Eligible"
    present_count=0
    for day in attendance:
        if day.upper()=='P':
            present_count +=1
    percentage=(present_count/len(attendance)) * 100
    print("Percentage", percentage)
    if percentage>=75:
        return "Eligible"
    else:
        return "Not Eligible"
if __name__=="__main__":
    records=["P", "P", "A", "P", "P"]
    print(check_eligibility(records))
```

6. Password Strength Checker

```python
def check_password(password):
    if len(password)<8:
        return "Weak"
    inc_digit=inc_special=False
    for char in password:
        if char.isdigit():
            inc_digit = True
        if char in "@#$":
            inc_special = True
    return "Strong Password" if inc_digit and inc_special else "Weak Password"
print(check_password("Admin@1234"))
```
